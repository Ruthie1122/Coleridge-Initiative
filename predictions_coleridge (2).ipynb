## Built a sep CNN

## Classifiers Models :
### There, we'll use classical classification models to predict text data from bag-of-words in our dataset. We will use 3 models: Random Forest, MultinomialNB and SGDClassifiers. To use these models, we'll transform features and targets into vectors, using TfidfVectorizer during the dataset extraction.

```python
X_train_cl = X_train.to_numpy()
X_test_cl = X_test.to_numpy()
y_train_cl = y_train.to_numpy()
y_test_cl = y_test.to_numpy()
```

```python
import string
from nltk.corpus import stopwords

stopwords_list  = list(string.punctuation) + stopwords.words('english')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(analyzer = 'word', token_pattern = r'\w{1,}',stop_words = stopwords_list)

X_train_vec = vectorizer.fit_transform(X_train_cl)
X_test_vec = vectorizer.transform(X_test_cl)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

top_charac = 20000

selector = SelectKBest(f_classif, k = min(top_charac, X_train_vec.shape[1])) #en fixant le nombre maximal de caractéristiques à sélectionner comme le minimum entre `top_k` et le nombre de caractéristiques présent dans le train set.
selector.fit(X_train_vec , y_train_cl)

X_train_vec = selector.transform(X_train_vec).astype('float32')
X_test_vec = selector.transform(X_test_vec).astype('float32')
```

```python
X_train_vec.shape
```

#### SelectKBest() selects the k best features using the ANOVA test, thus reducing the dimensionality of the dataset. This can help improve the quality of machine learning models by focusing on the most relevant features.

```python
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()

encoder.fit(list(y_train_cl) + list(y_test_cl) )
y_train_encoded = encoder.transform(y_train_cl)
y_test_encoded = encoder.transform(y_test_cl)
```
